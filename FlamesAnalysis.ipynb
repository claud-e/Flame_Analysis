```python
import pandas as pd

#Path to data
file_to_load="Files.csv/Data.csv"
```

```python
#New DataFrame
flame_data=pd.read_csv(file_to_load)
#Clean the DataFrame
flame_df=flame_data.iloc[:,:-2]

flame_df=flame_df.set_index("ID")
flame_df.describe()
```

```python
flame_votes=flame_df.groupby("Vote").count()
flame_votes
```

```python
plot=flame_votes.plot.pie(title="Percentage of votes",autopct='%1.1f%%')
```

```python
split_df=flame_df.groupby(["Vote","Gender"])["Gender"].count()
labels=["No,F","No,M","Yes,F","Yes,M"]
colors=["salmon","coral","skyblue","deepskyblue"]

split_df
```

```python
plot=split_df.plot.pie(title= "Gender split",autopct='%1.1f%%',subplots=True,labels=labels,colors=colors)
```

```python
flame_genders=flame_df.groupby("Gender").count()
flame_genders
plot=flame_genders.plot.pie(title= "Voters by gender",autopct='%1.1f%%',subplots=True)
```
